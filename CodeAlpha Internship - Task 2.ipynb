```python
dataset = data.values  # Now this should be (2515, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)
```

```python
training_data_len = int(len(scaled_data) * 0.8) # (80% of the dataset)

# Split the data
train_data = scaled_data[0:training_data_len]
```

```python
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])  # 60 previous values
    y_train.append(train_data[i, 0])       # the 61st value (target)

# Convert to numpy arrays
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, SpatialDropout1D

# Create a Sequential model
model = Sequential()

# First LSTM layer with return_sequences set to True
model.add(LSTM(64, return_sequences=True, input_shape=(x_train.shape[1], 1)))

# Adding SpatialDropout1D to prevent overfitting
model.add(SpatialDropout1D(0.3))

# Second LSTM layer without returning sequences (2D output)
model.add(LSTM(64, return_sequences=False))

# Output layer with a single neuron for prediction
model.add(Dense(1))

# Compile the model using 'adam' optimizer and 'mse' loss function
model.compile(optimizer='adam', loss='mse')

# Fit the model with training data
model.fit(x_train, y_train, epochs=25, batch_size=64)
```
